## Age Distribution and Customer Age Groups vs Customer Count and Churn Percentage

```python
import plotly.express as px

px.histogram(
    Bank_Churners_df,
    x="Customer_Age",
    nbins=20,
    title="Distribution of Customer Age",
    color_discrete_sequence=["#636EFA"]
)
```

```python
Bank_Churners_df["Age_Group"] = pd.cut(
    Bank_Churners_df["Customer_Age"],
    bins=[25, 35, 45, 55, 65, 75],
    labels=["25–35", "35–45", "45–55", "55–65", "65+"]
)
```

```python
age_summary = Bank_Churners_df.groupby("Age_Group").agg(
    Total_Customers=("Attrition_Flag", "count"),
    Churn_Percentage=("Attrition_Flag",
                      lambda x: (x == "Attrited Customer").mean() * 100)
).reset_index()
age_summary
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# عدد العملاء
fig.add_bar(
    x=age_summary["Age_Group"],
    y=age_summary["Total_Customers"],
    name="Number of Customers",
    marker_color="#636EFA"
)

# نسبة churn
fig.add_scatter(
    x=age_summary["Age_Group"],
    y=age_summary["Churn_Percentage"],
    name="Churn Percentage (%)",
    mode="lines+markers",
    yaxis="y2",
    line=dict(color="#EF553B", width=3)
)

fig.update_layout(
    title="Customer Age Groups vs Customer Count and Churn Percentage",
    xaxis_title="Age Group",
    yaxis=dict(title="Number of Customers"),
    yaxis2=dict(
        title="Churn Percentage (%)",
        overlaying="y",
        side="right"
    )
)

fig.show()
```

Younger customer segments tend to show relatively `higher churn rates` **compared to** `older customers`, suggesting that customer maturity and longer-term financial engagement may contribute to stronger retention behavior.

## Credit Limit Distribution and Churn Percentage VS Credit Limit Groups

```python
px.histogram(
    Bank_Churners_df,
    x="Credit_Limit",
    nbins=30,
    color="Attrition_Flag",
    title="Distribution of Credit Limit",
)
```

```python
Bank_Churners_df["Credit_Limit_Group"] = pd.qcut(
    Bank_Churners_df["Credit_Limit"],
    5,
    labels=["Very Low", "Low", "Medium", "High", "Very High"]
)
```

```python
churn_credit= (
    Bank_Churners_df.groupby('Credit_Limit_Group')['Attrition_Flag']
    .value_counts(normalize= True)
    .mul(100)
    .rename('Churn Percentage')
    .reset_index()
)
churn_credit= churn_credit[churn_credit['Attrition_Flag']== 'Attrited Customer']
churn_credit
```

```python
fig = go.Figure()

fig.add_scatter(
    x=churn_credit["Credit_Limit_Group"],
    y=churn_credit["Churn Percentage"],
    name="Churn Percentage (%)",
    mode="lines+markers",
    line=dict(color="#EF553B", width=3)
)

fig.update_layout(
    title="Churn % VS Credit Limit Group",
    xaxis_title="Credit Limit Group",
    yaxis_title="Churn Percentage (%)"
)

fig.show()
```

```python
Income_Order=[
    'Less than $40K',
    '$40K - $60K',
    '$60K - $80K',
    '$80K - $120K',
    '$120K +',
    'Unknown'
]
fig= px.histogram(
    Bank_Churners_df,
    x= 'Income_Category',
    category_orders={"Income_Category": Income_Order},
    title=('Income Category Distribution'),
    color= 'Attrition_Flag'
)
fig.show()
```

```python
churn_income = (
    Bank_Churners_df.groupby('Income_Category')['Attrition_Flag']
    .value_counts(normalize=True)
    .mul(100)
    .rename('Churn_Percentage')
    .reset_index()
)
```

```python
churn_income = churn_income[
    churn_income['Attrition_Flag'] == 'Attrited Customer'
]
```

```python
fig = px.bar(
    churn_income,
    x='Income_Category',
    y='Churn_Percentage',
    category_orders={'Income_Category': Income_Order},
    title='Churn % VS Income Category',
    color_discrete_sequence=['#EF553B']
)

fig.show()
```

## Churn Percentage VS Card Category

Customers holding `Platinum` cards exhibit the **highest churn rate among all card categories**, `indicating a potential retention risk within this premium segment. `

This may suggest unmet expectations regarding service value, benefits, or engagement compared to other card tiers. 

In contrast, `Silver` and `Blue` cardholders show relatively **lower churn rates**, reflecting more stable retention behavior.

These findings highlight the importance of reviewing benefit structures and engagement strategies specifically for high-tier cardholders to improve customer loyalty.

## Inactive 12 Months Distribution and Inactive Months vs Customer Count and Churn Percentage

```python
px.histogram(
    Bank_Churners_df,
    x="Months_Inactive_12_mon",
    title="Distribution of Inactive Months in the Last 12 Months",
    color_discrete_sequence=["#636EFA"]
)
```

The majority of customers fall `within 2–3 inactive months`, while extended inactivity `(4+ months)` is relatively rare and may signal potential disengagement risk.

Although most customers remain `inactive for only 2–3 months`, higher inactivity durations should be monitored closely as they may represent early warning signals of potential customer attrition.

```python
inactive_summary = Bank_Churners_df.groupby('Months_Inactive_12_mon').agg(
    Total_Customers=('Attrition_Flag','count'),
    Churn_Percentage=('Attrition_Flag', lambda x: (x=='Attrited Customer').mean()*100)
).reset_index()

inactive_summary
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Bar chart لعدد العملاء
fig.add_bar(
    x=inactive_summary["Months_Inactive_12_mon"],
    y=inactive_summary["Total_Customers"],
    name="Number of Customers",
    marker_color="#636EFA"
)

# Line chart لنسبة churn
fig.add_scatter(
    x=inactive_summary["Months_Inactive_12_mon"],
    y=inactive_summary["Churn_Percentage"],
    name="Churn Percentage (%)",
    mode="lines+markers",
    yaxis="y2",
    line=dict(color="#EF553B", width=3)
)

# إعداد المحور الثاني
fig.update_layout(
    title="Inactive Months vs Customer Count and Churn Percentage",
    xaxis_title="Months Inactive (Last 12 Months)",
    yaxis=dict(title="Number of Customers"),
    yaxis2=dict(
        title="Churn Percentage (%)",
        overlaying="y",
        side="right"
    ),
    legend=dict(x=0.01, y=0.99)
)

fig.show()
```

Although customers with `zero inactive months` show a relatively `high churn percentage`, **the small sample size in this group reduces reliability**. 

Churn probability increases noticeably `after 3 months of inactivity`, indicating that extended inactivity is a `strong early warning signal of potential customer attrition` and should trigger proactive retention strategies.
